**Complex numbers**

In Python, a complex number is a special type of number that has both a real part and an imaginary part. A complex number is written in the form a + bj, where a is the real part, b is the imaginary part, and j is the imaginary unit (which is the square root of -1).

```python
a = 2 + 3j # a+bj 
print(a)
type(a)
```

```python
# Adding two complex numbers
result = (2 + 3j) + (1 - 2j)
print(result) 
```

**Primitive Data Types**

Definition: The basic, built-in types of data in Python. These store a single value and are immutable (cannot be changed after creation).

**Non-Primitive Data Types**

Definition: More complex types that can store multiple values or structured data. They are usually mutable (can be changed).


**Primitive data types** are basic data types directly supported by the programming language (e.g., integers, floats, and booleans), while **non-primitive data types** are more complex and are derived from primitive types (e.g., lists, tuples, and dictionaries).

**Reserved keywords**
Reserved keywords cannot be used as identifiers (variable names, function names, etc.) because they have a special meaning in the Python language. For example, the keyword if is used to start a conditional statement, and the keyword for is used to start a loop. If you tried to use one of these keywords as a variable name, you would get an error.

**Swap 2 variables**

```python
#wrong approach
a = 2    # desired result a = 3 and b = 2
b = 3
a = b    # a = 3
b = a    # b = 3
print(a,b)
```

```python
# swapping using temp variable
a = 2    # desired result a = 3 and b = 2
b = 3
c = a    # c = 2
a = b    # a = 3
b = c    # b = 2
print(a,b)
```

```python
# Swapping 2 variables without using temp variable

a = 2 
b = 3
a,b = b,a
print(a,b)
```

```python
# Swapping 2 variables with logic
a = 2 
b = 3

a = a+b       # a = 5     --> 2+3 = 5
b = a-b       # b = 2     --> 5-3 = 2
a = a-b       # a = 3     --> 5-2 = 3
```
